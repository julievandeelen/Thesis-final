```python
#import relvant packages
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

from ema_workbench import ema_logging, load_results
from ema_workbench.analysis import (get_ex_feature_scores,
                                    RuleInductionType)

ema_logging.log_to_stderr(level=ema_logging.INFO)

#importing the data
fn =  './data/data_kde7.tar.gz'

x, outcomes = load_results(fn)
#x = x.drop(['model', 'policy'], axis=1)
```

```python
#get the scores for outcome atmosphere


y = outcomes['Atmospheric C']


atm_scores = []

top_5 = set()
for i in range(2, y.shape[1], 8):
    data = y[:, i]
    scores = get_ex_feature_scores(x, data,
                                   mode=RuleInductionType.REGRESSION)[0]

    top_5 |= set(scores.nlargest(3, 1).index.values)   
    scores = scores.rename(columns={1:outcomes['TIME'][0, i]})
    atm_scores.append(scores)
    
atm_scores = pd.concat(atm_scores, axis=1, sort=False)
atm_scores = atm_scores.loc[top_5, :]

```

```python
#get scores for outcome biomass


y = outcomes['Biomass mesopelagic fish']

bio_scores = []

top_5 = set()
for i in range(2, y.shape[1], 8):
    data = y[:, i]
    scores = get_ex_feature_scores(x, data,
                                   mode=RuleInductionType.REGRESSION)[0]

    top_5 |= set(scores.nlargest(3, 1).index.values)   
    scores = scores.rename(columns={1:outcomes['TIME'][0, i]})
    bio_scores.append(scores)
    
bio_scores = pd.concat(bio_scores, axis=1, sort=False)
bio_scores = bio_scores.loc[top_5, :]

```

```python
#get scores for outcome food provision


y = outcomes['Food provision by MF']


food_scores = []

top_5 = set()
for i in range(2, y.shape[1], 8):
    data = y[:, i]
    scores = get_ex_feature_scores(x, data,
                                   mode=RuleInductionType.REGRESSION)[0]

    top_5 |= set(scores.nlargest(3, 1).index.values)   
    scores = scores.rename(columns={1:outcomes['TIME'][0, i]})
    food_scores.append(scores)
    
food_scores = pd.concat(food_scores, axis=1, sort=False)
food_scores = food_scores.loc[top_5, :]

```

```python
#scores for outcome atmosphere


y = outcomes['Total vertical migration']


vm_scores = []

top_5 = set()
for i in range(2, y.shape[1], 8):
    data = y[:, i]
    scores = get_ex_feature_scores(x, data,
                                   mode=RuleInductionType.REGRESSION)[0]

    top_5 |= set(scores.nlargest(3, 1).index.values)   
    scores = scores.rename(columns={1:outcomes['TIME'][0, i]})
    vm_scores.append(scores)
    
vm_scores = pd.concat(vm_scores, axis=1, sort=False)
vm_scores = vm_scores.loc[top_5, :]

```

```python
atm_plot=atm_scores
bio_plot=bio_scores
food_plot=food_scores
vm_plot=vm_scores
```

```python
#plotting the feature scroing over time for the atmoshpere
title='atmospheric C'
sns.heatmap(atm_plot, cmap='viridis')
fig = plt.gcf()
ax = fig.get_axes()
plt.xticks(np.arange(0, 1440, 159.5),[2010,2020,2030,2040,2050,2060,2070,2080,2090,2100])


fig.set_size_inches(15,5)
fig.suptitle('Feature scoring over time for uncertainties most influential for '+title)
shorttitle = title.replace(" ","")
plt.savefig('./Figures/feature_atm_last.png',bbox_inches="tight")
plt.show()
```

```python
#plotting the feature scroing over time for the biomass
title='mesopelagic biomass'
sns.heatmap(bio_plot, cmap='viridis')
fig = plt.gcf()
ax = fig.get_axes()
plt.xticks(np.arange(0, 1440, 159.5),[2010,2020,2030,2040,2050,2060,2070,2080,2090,2100])

fig.set_size_inches(15,5)
fig.suptitle('Feature scoring over time for uncertainties most influential for '+title)
shorttitle = title.replace(" ","")
plt.savefig('./Figures/feature_bio_last.png',bbox_inches="tight")
plt.show()
```

```python
#plotting the feature scroing over time vertically migrated carbon
title='vertically migrated carbon'
sns.heatmap(vm_plot, cmap='viridis')
fig = plt.gcf()
ax = fig.get_axes()
plt.xticks(np.arange(0, 1440, 159.5),[2010,2020,2030,2040,2050,2060,2070,2080,2090,2100])

ax[0].set_xlabel('Time (Years)')
fig.set_size_inches(15,5)
fig.suptitle('Feature scoring over time for uncertainties most influential for '+title)
shorttitle = title.replace(" ","")
plt.savefig('./Figures/feature_vm_last.png',bbox_inches="tight")
plt.show()
```
